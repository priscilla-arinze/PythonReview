## Using `re.search()` instead of `find()`

Normal method:

```python
handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()
    
    if line.find('seabear') >= 0: # gives you -1 (returns false) if keyword is not found
        print(line)
```

Regular expressions method (using above import statement):

```python
# import re (already imported above)

handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()

    if re.search(r'seabear', line): # 'r' for raw string
        print(line)
```

## Using `re.search()` instead of `startswith()`

Normal method:

```python
handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()
    
    if line.startswith('Squidward:'): 
        print(line)
```

Regular expressions method (using above import statement):

```python
handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()
    
    if re.search(r'^Squidward:', line):
        print(line)
```

## Wildcard Characters

The dot (.) matches any character

The asterisk (*) is for a certain character/text to appear ZERO or more times

The plus sign (+) is for a certain character/text to appear ONE or more times

### Example 1:

```python
from IPython.display import Image
from IPython.core.display import HTML 
Image(filename="../aa_Screenshots/regex-wildcard-example.png")
```

```python
handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()
    
    if re.search(r'seabear.*Squidward', line):
        print(line)
```

### Example 2:

```python
from IPython.display import Image
from IPython.core.display import HTML 
Image(filename="../aa_Screenshots/regex-wildcard-example2.png")
```

## Matching & Extracting Data

`re.search()`: Allows you to see whether there are matches and use them for conditional statements for example <br>
`re.findall()`: Returns all matches

__[0-9]+__: Return any digit between 0 and 9 ONE or more times

```python
txt = 'My 3 favorite numbers are 7, 21, and delaware'

print(re.findall(r'[0-9]+', txt)) # returns a list of strings
```

```python
print(re.findall(r'[aeiou]+', txt))
```

```python
print(re.findall(r'[AEIOU]+', txt)) #CASE-SENSITIVE
```

## Greedy Matching

Greedy matching for regex is basically finding the character/keyword match within a line and continuing on to the end of the line instead of stopping at the end of the match.

Example from above:

*Will display the entire line instead of just stopping at "Squidward:"*

```python
handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()
    
    if re.search(r'^Squidward:.+', line):
        print(re.findall(r'^Squidward:.+', line))
```

## Non-Greedy Matching

Opposite of above; only displays exact matches instead of whole line

Add question mark (?) right BEFORE the final character/txt of the string

```python
handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()
    
    if re.search(r'^Squidward:.+?', line):
        print(re.findall(r'^Squidward:.+?', line))
```

## Fine-Tuning String Extraction

Example: Extracting all actions/events described in episode that are enclosed between parenthesis (need escape characters)

*Without fine-tuning*

```python
handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()
    
    if re.search(r'\(.+\)', line):
        print(re.findall(r'\(.+\)', line))
```

```python
subject = 'From birds@governmentdrones.com Sat Jul 2 01:10:16 2022; From cottoncandy@icecream.gov Fri Jun 10 15:07:35 2022'

print(re.findall(r'@.+', subject))
```

*With fine-tuning*

```python
handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()
    
    if re.search(r'\(.*?\)', line):
        print(re.findall(r'\(.*?\)', line))

# if there are two sets of parentheses within the same line, they will be comma separated in the same list
```

```python
subject = 'From birds@governmentdrones.com Sat Jul 2 01:10:16 2022; From cottoncandy@icecream.gov Fri Jun 10 15:07:35 2022'

print(re.findall(r'\S+@\S+', subject))
```

### More fine-tuning with parentheses

Parentheses tell you where to start & stop for the string extraction

Example: Extracting all actions/events described in episode that are enclosed between parenthesis (need escape characters) for Squidward's lines ONLY

```python
handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()
    
    if re.search(r'^Squidward:.*\(', line):
        print(re.findall(r'(\(.*?\))', line))
```

Only searching for email addresses that begin with "From":

```python
subject = """From birds@governmentdrones.com Sat Jul 2 01:10:16 2022;
From cottoncandy@icecream.gov Fri Jun 10 15:07:35 2022
refrigerator@worlddomination.com Fri Jun 10 5:37:02 2022
From macaroni@andcheese.org Sat Jul 2 00:10:35 2022
"""

print(re.findall(r'From (\S+@\S+)', subject))
```

## String Parsing

Without regular expressions using `find()`:

```python
subject = 'From birds@governmentdrones.com Sat Jul 2 01:10:16 2022'

at_sign_pos = subject.find('@') # stores index of at (@) symbol
print(at_sign_pos) # 10

# NOTE: find() has the following parameters: value (required), starting index (optional), ending index (optional)
space_pos = subject.find(' ', at_sign_pos) # find 1st space, starting from the at sign
print(space_pos) #31

email_domain = subject[at_sign_pos+1 : space_pos]
print(email_domain)
```

Without regular expressions using `split()`:

```python
indiv_words = subject.split()
print("List of split words:", indiv_words) # list of individual words from original string, split at each whitespace

email = indiv_words[1] # email address
print("Email Address", email)

email_pieces = email.split('@') # split at each '@' sign instead of splitting by default at each whitespace
print("List of each part of email address:", email_pieces)

print("\n")

print("Domain Address:", email_pieces[1])
```

With regular expressions:

```python
from IPython.display import Image
from IPython.core.display import HTML 
Image(filename="../aa_Screenshots/regex-wildcard-example3.png")
```

*Note: The caret character (^) by itself is to find matches at the beginning of text; <br> The caret character (^) inside brackets ([]) is to find any character EXCEPT characters after the caret inside the brackets*

```python
regex = re.findall(r'@([^ ]*)', subject)

print(regex)
```

Fine-tuned regular expression version (only find email addresses where the line starts with "From"):

```python
subject = """From birds@governmentdrones.com Sat Jul 2 01:10:16 2022;
From cottoncandy@icecream.gov Fri Jun 10 15:07:35 2022
refrigerator@worlddomination.com Fri Jun 10 5:37:02 2022
From macaroni@andcheese.org Sat Jul 2 00:10:35 2022
"""

print(re.findall(r'From .*@([^ ]*)', subject))
```

## Mini-Exercise: Find average & median hourly/annual salaries for Texas

Source: [U.S. Bureau of Labor Statistics](https://www.bls.gov/oes/current/oes_nat.htm)

```python
from numpy import median

handle = open('BLS-Texas-May2021-WageStatistics.txt')

hourly_salaries_list = list()
annual_salaries_list = list()

for line in handle:
    line = line.rstrip()
    
    hourly_salaries = re.findall('(\S+\.\S+)', line)
    annual_salaries = re.findall('(\S+,\S+)', line)

    # Each number is stored in its own individual list

    # avoid blank lines or lines with multiple hourly decimal numbers
    if len(hourly_salaries) == 1:
        hr_num = float(hourly_salaries[0])
        hourly_salaries_list.append(hr_num)

        avg_hourly_salary = (sum(hourly_salaries_list) / len(hourly_salaries_list))
        median_hourly_salary = median(hourly_salaries_list)


    # avoid blank lines or lines with multiple annual salary numbers with commas
    if len(annual_salaries) == 1:
        annual_salaries_num = int(annual_salaries[0].replace(',', ''))
        annual_salaries_list.append(annual_salaries_num)

        avg_annual_salary = (sum(annual_salaries_list) / len(annual_salaries_list))
        median_annual_salary = median(annual_salaries_list)
    
print('Average Hourly:', "{:.2f}".format(avg_hourly_salary), "; Median Hourly:", "{:.2f}".format(median_hourly_salary))
print('Average Annual:', "{:.2f}".format(avg_annual_salary), "; Median Annual:", "{:.2f}".format(median_annual_salary))
```
