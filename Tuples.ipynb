# Tuples

Very similar to lists; surrounded by parenthesis instead of square brackets

__Difference to lists: Tuples are IMMUTABLE__ (cannot alter [assign new values] after creation); only allows unique keys

**********

Benefits: tuples are more efficient than lists, take up less storage, quicker to access, great for temporary usage

Disadvantages: Unlike lists, tuples cannot be sorted, cannot append items at the end, cannot reverse order

## Tuples & Double Assignment Ability

Able to put a tuple of variables on the left-hand side of assignment statement (parentheses not required)

```python
(num1, num2) = (9, 81)

print(num2)
```

```python
year, make, model = (2003, 'Honda', 'Civic')

print(make)
```

__SWAP 2 VARIABLES EASILY__:

```python
x = 10
y = 20

print(x, y)

x, y = y, x # swapped using tuple double assignment

print(x, y)
```

## Tuples & Dictionaries

```python
food = dict()

food['rice'] = 'delicious'
food['fish'] = 'spectacular'
food['mustard'] = 'satisfactory'

for (key, val) in food.items():
    print(key, val)
```

Prints a list of (key, value) tuples:

```python
tups = food.items()
print(tups)
```

## Tuples are Comparable 
Left to right comparison, returns boolean

Comparison stops until True, otherwise final result is False if ALL comparisons are False

```python
# 0 is NOT greater than 5 ----> FALSE
# 1 = 1, SKIP
# 2 = 2, SKIP 
(0, 1, 2) > (5, 1, 2)
```

```python
# 0 = 0, SKIP
# 1 IS less than 3 ----> TRUE (comparison stops here)
# *NOT EVALUATED*: previous comparison is True; 500000 is NOT less than 4
(0, 1, 500000) < (0, 3, 4)
```

```python
# each letter is compared from left to right; m comes AFTER l (greater than), not before (less than)
('Refrigerator', 'Samantha') < ('Refrigerator', 'Sally')
```

```python
# v comes after b (greater than) --> True
('Ohio', 'Nevada') > ('Ohio', 'Nebraska')
```

## Sorting

Tuples themselves can't be sorted, but can be sorted when converted to a list of items

```python
a = {'bob':37, 'abigail':26, 'priscilla':23, 'brick':78, 'squidward':974}

a.items()
```

```python
sorted(a.items())
```

```python
for k, v in sorted(a.items()):
    print(k, "---", v)
```

Sorting by values instead of keys:

*basically will have to switch the key, value order using a list*

```python
temp = list()

for k, v in a.items():
    temp.append( (v,k) ) # switching key, value order

print(temp)
```

```python
temp = sorted(temp, reverse=True) # will sort in descending order

print(temp)
```

Condensed version of sorting tuples:

```python
print( sorted( [ (v,k) for k,v in a.items() ] ) )
```
