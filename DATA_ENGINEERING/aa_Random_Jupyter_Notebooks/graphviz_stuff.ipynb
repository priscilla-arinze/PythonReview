```python
import graphviz

g = graphviz.Digraph('test-graph', comment='A test graph', strict=True)

g.edge("A", "B", color="blue")
g.edge("A", "B", color="red")
g.edge("A", "B", color="blue")
g.edge("A", "B", color="green")
g.edge("B", "A", color="purple")

c1 = 'A'
c2 = 'B'

if "\t{c1} -> {c2}".format(c1=c1, c2=c2) in g.source:
    g.edge("A", "B", color="red")

g
```
