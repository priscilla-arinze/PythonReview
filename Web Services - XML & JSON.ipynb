## XML

"eXtensible Markup Language"; elements/nodes (< ... >) can be named whatever, unlike HTML with its set standard elements (<'p'>, <'header'>)

```python
from IPython.display import Image
from IPython.core.display import HTML 
Image(filename="aa_Screenshots/xml-basics.png", width=700)
```

*Note: XML is not as forgiving as HTML with syntax errors; cannot parse at all with syntax errors*

```python
import xml.etree.ElementTree as ET # XML parser; think of XML like a tree (elements represent parent & child nodes)

data = '''
<person>
    <name>Priscilla</name>
    <phone type="intl"> +1 123 456 5555 </phone>
    <email hide="yes"/>
</person>
'''

tree = ET.fromstring(data)

# tree.find(< ELEMENT NAME >).[method]
print('Name:', tree.find('name').text)
print('Attr:', tree.find('email').get('hide'))
```

```python
import xml.etree.ElementTree as ET # XML parser; think of XML like a tree (elements represent parent & child nodes)

data2 = '''
<users>
    <people>
        <person x="2">
            <id>001</id>
            <name>Priscilla</name>
        </person>
        <person x="7">
            <id>009</id>
            <name>Squidward</name>
        </person>
    </people>
</users>
'''

tree = ET.fromstring(data2)


 # within users, find all the child elements under <people>
 # stored as a list of elements only, not contents inside tags
childnodes = tree.findall('people/person') 
print("Child nodes:", childnodes)
print('User count:', len(childnodes), "\n") # counts number of elements

for item in childnodes:
    # tree.find(< ELEMENT NAME >).[method]
    print('Name:', item.find('name').text)
    print('ID', item.find('id').text)
    print('Attribute', item.get("x"))
    print("\n")
```

## JSON

Javascript Object Notation; very similar syntax to Python dictionaries & nested "lists"

```python
import json

data = '''
{
    "name" : "Priscilla",
    "phone" : {
        "type" : "intl",
        "number" : "+1 123 456 5555"
    },
    "email" : {
        "hide" : "yes"
    }
}
'''

info = json.loads(data) # JSON parser
# indexing has very similar syntax to Python dictionaries
print('Name:', info["name"])
print('Hide:', info["email"]["hide"]) # nested list representation

```

```python
import json

# list with multiple (2) dictionaries/objects
data2 = ''' [
    {
        "id" : "001",
        "x" : "2",
        "name" : "Priscilla"
    },
    {
        "id" : "009",
        "x" : "7",
        "name" : "Squidward"
    }
]'''

info = json.loads(data2) # JSON parser
print('User count:', len(info), "\n")

for item in info:
# indexing has very similar syntax to Python dictionaries
    print('Name:', item["name"])
    print('ID:', item["id"])
    print('Attribute:', item["x"])
    print("\n")

```
