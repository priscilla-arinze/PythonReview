# Parse HTML with Beautiful Soup

```python
url = "https://coppermind.net/wiki/Navani_Kholin"

r = requests.get(url)

html_doc = r.text

soup = BeautifulSoup(html_doc)

pretty_soup = soup.prettify()

print(pretty_soup)
```

```python
print(soup.title)
```

```python
print(soup.get_text())
```
