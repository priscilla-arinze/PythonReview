```python
formula_columns = {
    'Directory': ['I'],
    'File': ['J']
}
```

```python
## Parsing FileName column
for row in range(2, totalrows_rawdata+1):
    # FileName column
    current_cell_filename = ws['A'+str(row)].value
    
    # Directory column
    strmatch_dir = r'(^.*)/'
    directory = re.findall(strmatch_dir, current_cell_filename)[0]
    formula_columns['Directory'].append(directory)

    # File column
    strmatch_file = r'^.*/(.*$)'
    file = re.findall(strmatch_file, current_cell_filename)[0]
    formula_columns['File'].append(file)


    
#wb.save("output.xlsx")
```

```python
print(len(formula_columns['Directory']))
```

```python
import pandas as pd

df = pd.DataFrame(formula_columns)
```

```python
# Drop first row
df.drop(index=df.index[0], axis=0, inplace=True)
```
