# Network Programming - Web Scraping & Parsing HTML

Using BeautifulSoup

```python
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
```

```python
## For websites with expired SSL certificates:
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

## User input for URL
url = input('Enter URL: ') # Must be the full url (including the http part); e.g. https://simple.wikipedia.org/wiki/Nathan_MacKinnon

## Open & read web page contents of above URL
html = urllib.request.urlopen(url).read()

## Parse HTML with BeatifulSoup
soup = BeautifulSoup(html, 'html.parser')

## Find all anchor <a> tags & print its href URL
tags = soup('a')

for tag in tags:
    print(tag.get('href', None)) # get the href url only from a tags
```

## Goodreads: Books Read in 2024

```python
url = "https://www.goodreads.com/review/list/12153793-priscilla-arinze?read_at=2024"

html = urllib.request.urlopen(url).read()

soup = BeautifulSoup(html, 'html.parser')

books_read_2024 = soup.find_all("td", attrs={"class": "field title"})
books_read_2024
```

```python
for title in books_read_2024:
    print(title.div.a.get("title"))
```
