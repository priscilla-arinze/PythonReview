# Writing Files

## Regular write to file

__NOTE: "w" overwrites the entire file__

```python
with open("write_example.txt", "w") as file1:
    file1.write("spongebob\n")
    file1.write("patrick\n")
    file1.write("patrick\n")
```

## From list of lines

```python
lines = ["spongebob\n", "patrick\n", "squidward\n"]

with open("write_example2.txt", "w") as file2:
    for line in lines:
        file2.write(line)
```

## Appending to file

```python
with open("write_example.txt", "a") as file3:
    file3.write("sandy\n")
```
