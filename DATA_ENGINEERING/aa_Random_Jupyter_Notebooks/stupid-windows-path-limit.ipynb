```python
import subprocess
from subprocess import PIPE, Popen, CalledProcessError

def execute(cmd):
    popen = subprocess.Popen(cmd, stdout=subprocess.PIPE, universal_newlines=True)
    for stdout_line in iter(popen.stdout.readline, ""):
        yield stdout_line 
    popen.stdout.close()
    return_code = popen.wait()
    if return_code:
        raise subprocess.CalledProcessError(return_code, cmd)
```

```python
import os
import re

src_path = "D:/Users/priscillaw/babel_TEST_DIRS2"
dest_path = "D:/Users/priscillaw/COPY_TEST"

test_dirs = set()

for root, dirs, files in os.walk(src_path):
    for dir in dirs:
        full_path = os.path.normpath(os.path.join(root, dir))
        if '\\test' in full_path:
            full_path_re = re.findall(r"(^.*\\test).*$", full_path)[0]
            test_dirs.add(full_path_re)

for dir in test_dirs:
    if os.path.exists(dir):
        relname = os.path.relpath(dir, src_path)
        full_dest_path = os.path.normpath(os.path.join(dest_path, relname))
        ## Create directories in dest_path
        mkdir = "mkdir {dest}".format(dest=full_dest_path)
        print(mkdir)
        os.system(mkdir)
        
        ## Copy directory contents from src_path
        xcopy = "xcopy {src} {dest} /E/H".format(src=dir, dest=full_dest_path)
        print(xcopy)
        os.system(xcopy)

        ## Remove directories from src_path
        rmdir = "rmdir {src} /Q/S".format(src=dir)
        print(rmdir)
        os.system(rmdir)

        print("\n\n")

```
