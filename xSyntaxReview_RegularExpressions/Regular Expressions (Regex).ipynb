## Using `re.search()` instead of `find()`

Normal method:

```python
handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()
    
    if line.find('seabear') >= 0: # gives you -1 (returns false) if keyword is not found
        print(line)
```

Regular expressions method (using above import statement):

```python
# import re (already imported above)

handle = open('Spongebob_Episode_Transcript--The_Camping_Episode.txt')

for line in handle:
    line = line.rstrip()

    if re.search('seabear', line):
        print(line)
```
